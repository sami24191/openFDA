```python
import requests
import json
```

```python
# Base URL for the openFDA API
base_url = "https://api.fda.gov/device/registrationlisting.json"

# Parameters for the API request
# params = {
#     "search": "products.product_code:HQY",
#     "limit": 1
# }

params = {
    "search": "cgm",  # Search for 'aspirin'
    "limit": 100  # Limit results to 5 records
}

# Send the GET request to the API
response = requests.get(base_url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    # Process the data as needed
    print(json.dumps(data, indent=2))
else:
    print(f"Error: {response.status_code}")
    #print(response.text)
```

```python
import pandas as pd

# Extract relevant fields from the saved 'results'
device_data = []
for result in data:
    for product in result.get('products', []):
        device_data.append({
            # Fields from 'result' segment
            'proprietary_name': ', '.join(result.get('proprietary_name', [])),
            'establishment_type': ', '.join(result.get('establishment_type', [])),
            'registration_number': result.get('registration', {}).get('registration_number', ''),
            'status_code': result.get('registration', {}).get('status_code', ''),
            'initial_importer_flag': result.get('registration', {}).get('initial_importer_flag', ''),
            'reg_expiry_date_year': result.get('registration', {}).get('reg_expiry_date_year', ''),
            'name': result.get('registration', {}).get('name', ''),
            'city': result.get('registration', {}).get('city', ''),
            'state_code': result.get('registration', {}).get('state_code', ''),
            'iso_country_code': result.get('registration', {}).get('iso_country_code', ''),
            'zip_code': result.get('registration', {}).get('zip_code', ''),
            'postal_code': result.get('registration', {}).get('postal_code', ''),

            # Fields from 'us_agent' in 'registration'
            'us_agent_name': result.get('registration', {}).get('us_agent', {}).get('name', ''),
            'us_agent_business_name': result.get('registration', {}).get('us_agent', {}).get('business_name', ''),
            'us_agent_bus_phone_num': f"{result.get('registration', {}).get('us_agent', {}).get('bus_phone_area_code', '')}-{result.get('registration', {}).get('us_agent', {}).get('bus_phone_num', '')}",
            'us_agent_email_address': result.get('registration', {}).get('us_agent', {}).get('email_address', ''),
            'us_agent_city': result.get('registration', {}).get('us_agent', {}).get('city', ''),
            'us_agent_state_code': result.get('registration', {}).get('us_agent', {}).get('state_code', ''),
            'us_agent_iso_country_code': result.get('registration', {}).get('us_agent', {}).get('iso_country_code', ''),
            'us_agent_zip_code': result.get('registration', {}).get('us_agent', {}).get('zip_code', ''),
            'us_agent_postal_code': result.get('registration', {}).get('us_agent', {}).get('postal_code', ''),

            # Fields from 'owner_operator' in 'registration'
            'owner_operator_firm_name': result.get('registration', {}).get('owner_operator', {}).get('firm_name', ''),

            # Fields from 'products' segment
            'product_code': product.get('product_code', ''),
            'product_created_date': product.get('created_date', ''),
            'product_owner_operator_number': product.get('owner_operator_number', ''),
            'product_exempt': product.get('exempt', ''),

            # Fields from 'openfda' within 'products'
            'device_name': product.get('openfda', {}).get('device_name', ''),
            'medical_specialty_description': product.get('openfda', {}).get('medical_specialty_description', ''),
            'regulation_number': product.get('openfda', {}).get('regulation_number', ''),
            'device_class': product.get('openfda', {}).get('device_class', '')
        })

# Create a DataFrame from the extracted data
df = pd.DataFrame(device_data)

# Print the DataFrame
df
```
